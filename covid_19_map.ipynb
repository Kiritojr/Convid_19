```
import pandas as pd
import folium
from folium import plugins
```

```
dataset = pd.read_csv('covid-19-all.csv')
dataset.head()
```

```
latitude = dataset['Latitude'].fillna(0)

longitude = dataset['Longitude'].fillna(0)
```

```
cordenadas = []

for lat,lnt in zip(latitude,longitude):

    cordenadas.append([lat,lnt])
```

```
mapa = folium.Map()
mapa
```

```
mapa.add_child(plugins.HeatMap(cordenadas))        
mapa
```
