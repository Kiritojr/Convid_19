```
import pandas as pd
import plotly.offline as py
import plotly.graph_objs as go
```

```
dataset = pd.read_csv('covid-19-all.csv')

dataset.head(10)
```

```
dataset.drop(['Last Update', 'Latitude','Longitude','date_str'], axis=1)
```

```
confirmed = dataset['Confirmed']
date = dataset['date']
death = dataset['Deaths']
```

```
infected = []
a = z = 0
for count in confirmed.fillna(0):

    a = z + count
    infected.append(a)
    z = a
```

```
deaths = []
a = z = 0
for count in death.fillna(0):

    a = z + count
    deaths.append(a)
    z = a
```

```
data = [go.Scatter(x=date,
                   y=infected,
                   name='Infectados'), go.Scatter(x=date,
                                           y=deaths,
                                           name='Mortes')]

layout = go.Layout(title='casos de coronavírus no Mundo',
yaxis={'title':'casos confirmados'},
xaxis={'title':'data de observação'})
```

```
fig = go.Figure(data=data, layout=layout)

py.iplot(fig)
```
